```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
sys.path.append('../') 
from utils import Utils
utils_obj = Utils()
```

```python
df = utils_obj.load_data("train_data.csv")
#df = pd.read_csv("train_data.csv")
#print(df)
no_of_datapoints, no_of_unique_actions, no_of_unique_objects, no_of_unique_locations = utils_obj.get_data_stats(df)
print(f"No. of Datapoints {no_of_datapoints}")
print(f"No. of unique actions {no_of_unique_actions}")
print(f"No. of unique objects {no_of_unique_objects}")
print(f"No. of unique locations {no_of_unique_locations}")
```

```python
action_freq_dict = utils_obj.get_action_freq(df)
print(  list(action_freq_dict.keys())  )
plt.barh(list(action_freq_dict.keys()), list(action_freq_dict.values()) )

plt.title('action distribution')
plt.ylabel('Actions')
plt.xlabel('Frequency')
plt.show()
```

```python
object_freq_dict = utils_obj.get_object_freq(df)
plt.barh(list(object_freq_dict.keys()), list(object_freq_dict.values()) )
plt.title('Object Distribution')
plt.ylabel('Objects')
plt.xlabel('Frequency')
plt.show()
```

```python
location_freq_dict = utils_obj.get_location_freq(df)
plt.barh(list(location_freq_dict.keys()), list(location_freq_dict.values()) )
plt.title('location Distribution')
plt.ylabel('location')
plt.xlabel('Frequency')
plt.show()
```
